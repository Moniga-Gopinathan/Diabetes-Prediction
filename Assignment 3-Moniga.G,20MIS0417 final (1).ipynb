## 5. Handle the Missing values.

```python
#if we have null value in the numerical column then we have to replace them with Mean or Median
# else if the column that has the null value is categorical based then it should be replaced with Mode
#let us check for null values in the data set
data.isna()
```

```python
data.isnull().sum()#no null values are found in the given dataset so there is no need to handle them 
```

```python
valid_columns = data.select_dtypes(include=[np.number]).columns

# Fill missing values with mean of valid columns
data[valid_columns] = data[valid_columns].fillna(data[valid_columns].mean())
data
```

#### missing values -not found not necessary to handle them 

## 6. Find the outliers and replace the outliers

```python
sb.boxplot(data.area)# finding the outliers in the price column 
```

```python
data['price'] = np.where((data['price'] < lower_extreme['price']) | (data['price'] > upper_extreme['price']), data['price'].median(), data['price'])
data['price']
```

```python
plt.boxplot(data['price'])
```

### Outliers are changed

## 7. Check for Categorical columns and perform encoding.


#### Generally categorical column involves the column that are not numerical and involve some categorical division instead of the numerical values and numerical manipulation 

```python
#we have to focus on the column which are described as objects
#Label Encoder-they generally assign the numerical value for each category
#One hot handling -splits them to the number of classes
from sklearn.preprocessing import LabelEncoder
categorical_columns = data.select_dtypes(include=['object']).columns#converting the objects as they are categorical columns

# Perform encoding for each categorical column
for column in categorical_columns:
    
    label_encoder = LabelEncoder()
    data[column] = label_encoder.fit_transform(data[column])
print(data)#all the categorical columns are converted to numerical columns
```

## 14. Measure the performance using Metrics.

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
```

```python
#for linear regression
mse1=mean_squared_error(y_test,y_pred1)
mae1=mean_absolute_error(y_test,y_pred1)
rmse1=np.sqrt(mse1)
r2lr=r2_score(y_test,y_pred1)


print("Mean Absolute Error for linear regression(MAE):", mae1)
print("Mean Squared Error for linear regression (MSE):", mse1)
print("Root Mean Squared Error for linear regression(RMSE):", rmse1)
print("R-squared (R2)for linear regression:", r2lr)
```
